```python
def get_key(val, my_dict):
    for key, values in my_dict.items():
        if val in values:
            return key
 
    return "NotAvailable"
```

```python
def col_name(type_, date):
    nm = date+ ' ' + type_
    return nm
```

```python
#计算位置和站点的距离
def distance(curr_long, curr_lat, station_long, station_lat):
    curr_ = (curr_lat, curr_long)
    station_ = (station_lat, station_long)
 
    return (geodesic(curr_, station_).km)
    #return ((curr_long-station_long)**2+(curr_lat-station_lat)**2)**(1/2)
    #return (abs(curr_long-station_long)+abs(curr_lat-station_lat))
```

```python
def find_gps_sn(bus_id):
    return gps[gps['id']==bus_id]['gps_sn'].values[0]
```

```python
def pass_zhandian_with_time(date, bus_id, gps_data_all, site_gps):
    gps_sn = find_gps_sn(bus_id)
    current_df = gps_data_all[gps_data_all['gps串号']==int(gps_sn)]
    zhandian, time = find_stations(date, current_df, site_gps)
    print('date: ', date)
    print('经过站点: ', zhandian)
    print('经过时间: ', time)
    
    return zhandian, time
    
```

```python
def pass_area_with_time(date, bus_id, gps_data_all, site_gps):
    gps_sn = find_gps_sn(bus_id)
    current_df = gps_data_all[gps_data_all['gps串号']==int(gps_sn)]
    area, time = find_area(date, current_df, site_gps)
    print('date: ', date)
    print('经过区域: ', area)
    print('经过时间: ', time)
    
    return area, time
    
```

```python
#find nearest station
def possible_nearest_station(site_df, this_long, this_lat):
    time_format = "%Y-%m-%d %H:%M:%S"
    min_dist = float('inf')
    res = 'Not Found'

    for index, row in site_df.iterrows():
        site_long = row['经度']
        site_lat = row['纬度']
        dist = distance(this_long, this_lat, site_long, site_lat)
        if dist < min_dist: 
            
            min_dist = dist
            res = row['站点']
            
    return res     
```

```python
def find_stations(date, data, site_gps):
    pass_zhandian = []
    pass_time = []
    current_data_df = data[data['接收日期']==date]   
        
    #position info
    last_longitude = 0
    last_latitude = 0 
        
    last_nearest = ''
                  
    #iterate signals
    for index1, row1 in current_data_df.iterrows():
            
        nearest_= possible_nearest_station(site_gps, row1['经度'], row1['纬度'])
        if (nearest_ != last_nearest):
            pass_zhandian.append(nearest_)
            time_ = row1['接收时间'].time()
            pass_time.append(str(time_))
            last_nearest = nearest_
               
                
    return pass_zhandian, pass_time
```

```python
def find_area(date, data, site_gps):
    pass_zhandian = []
    pass_time = []
    current_data_df = data[data['接收日期']==date]   
        
    #position info
    last_longitude = 0
    last_latitude = 0 
        
    last_nearest = ''
                  
    #iterate signals
    for index1, row1 in current_data_df.iterrows():
            
        nearest_= possible_nearest_station(site_gps, row1['经度'], row1['纬度'])
        area = get_key(nearest_, MapArea)
        if area != "NotAvailable":
            nearest_ = area
        if (nearest_ != last_nearest):
            pass_zhandian.append(nearest_)
            time_ = row1['接收时间'].time()
            pass_time.append(str(time_))
            last_nearest = nearest_
               
                
    return pass_zhandian, pass_time
```

```python
def run_pattern_one_bus(bus_id, dates):
    pattern_df = pd.DataFrame()
    for each_day in dates:
        zhandian, time = pass_zhandian_with_time(each_day, bus_id, gps_info_up, site_gps)
        new_route_info = pd.DataFrame({
            col_name('站点', each_day): zhandian,
            col_name('时间', each_day): time
            })
        pattern_df = pd.concat([pattern_df, new_route_info], axis=1) 
            
    
    return pattern_df
```

```python
def run_pattern_one_bus_area(bus_id, dates):
    pattern_df = pd.DataFrame()
    for each_day in dates:
        area, time = pass_area_with_time(each_day, bus_id, gps_info_up, site_gps)
        new_route_info = pd.DataFrame({
            col_name('区域', each_day): area,
            col_name('时间', each_day): time
            })
        pattern_df = pd.concat([pattern_df, new_route_info], axis=1) 
            
    
    return pattern_df
```
